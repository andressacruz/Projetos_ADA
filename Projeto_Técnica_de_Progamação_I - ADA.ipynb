###Índice de Desenvolvimento da Educação Básica - IDEB

O IDEB  é uma métrica do governo brasileiro para avaliar a qualidade da educação. Ele é calculado com base nos resultados dos alunos em testes padronizados, como o SAEB (Sistema de Avaliação da Educação Básica), e nas taxas de promoção escolar. O índice é divulgado a cada dois anos e varia de 0 a 10. Quanto maior o valor, melhor o desempenho educacional. O IDEB é uma ferramenta fundamental para monitorar o progresso educacional no país e identificar áreas que precisam de melhorias. Os dados encontrados na tabela se referem ao período de 2005 até 2021.

### Ensino Médio

````python
# Seleciona apenas os dados referentes ao Ensino Médio.
selecao_ensino_medio = df_ideb.loc[df_ideb['ensino'] == 'medio' , :]

````

````python
# Seleciona os dados pelo tipo de rede

escolas_publicas = selecao_ensino_medio[selecao_ensino_medio['rede'] == 'publica']
escolas_privadas = selecao_ensino_medio[selecao_ensino_medio['rede'] == 'privada']
escolas_estaduais = selecao_ensino_medio[selecao_ensino_medio['rede'] == 'estadual']
total_escolas_EM = selecao_ensino_medio [selecao_ensino_medio['rede'] == 'total']
````

````python
#grafico de linha com as variáveis


eixo_x_publicas = escolas_publicas['ano']
eixo_y_publicas = escolas_publicas['taxa_aprovacao']

eixo_x_privadas = escolas_privadas['ano']
eixo_y_privadas = escolas_privadas['taxa_aprovacao']

eixo_x_estaduais = escolas_estaduais['ano']
eixo_y_estaduais = escolas_estaduais['taxa_aprovacao']

eixo_x_total = total_escolas_EM['ano']
eixo_y_total = total_escolas_EM['taxa_aprovacao']

plt.plot(eixo_x_publicas, eixo_y_publicas, label='Escolas Públicas')
plt.plot(eixo_x_privadas, eixo_y_privadas, label='Escolas Privadas')
plt.plot(eixo_x_estaduais, eixo_y_estaduais, label='Escolas Estaduais')
plt.plot(eixo_x_total, eixo_y_total, label='Total')

plt.xlabel('Ano')
plt.ylabel('Taxa de Aprovação')
plt.title('Taxa de Aprovação no Ensino Médio por "Rede"')

plt.legend()
plt.show()
````

- **Comparação por período:** Com as informações do gráfico acima foi possível notar que ao longo do período, houve crescimento nas taxas de aprovação de modo geral.  Na rede privada já se mantinha uma taxa elevada e seu crescimento foi menor comparado com as escolas Públicas e Estaduais, a qual possuem valores mais discrepantes.

- **Comparação entre variáveis:** As escolas privadas se mostraram um valor mais elevado comparado com as escolas da rede público e estaduais e a média total.
